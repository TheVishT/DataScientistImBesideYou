```python
# GAZE DATA
```

```python
# Understanding the data 
```

```python
gaze_df=pd.read_csv("emotion_data/1/gaze.csv") # Loading the gaze data of 1st student
gaze_df=gaze_df.drop(gaze_df.index[-1]) # Removing the last row of gaze of student 1
```

```python
for i in range(2,11): # Merging the gaze data of all students in gaze_df
    df2=pd.read_csv("emotion_data/"+str(i)+"/gaze.csv")
    gaze_df=pd.merge(gaze_df,df2,how='outer')
```

```python
gaze_df.head()
```

```python
gaze_df.shape
```

```python
gaze_df.describe()
```

```python
# Cleaning the data
```

```python
gaze_df.columns
```

```python
gaze_df=gaze_df.drop(['movie_id','image_seq','blink'],axis=1) # These columns don't contribute to analysis
gaze_df
```

```python
# Relationship Analysis
```

```python
correlation_matrix = gaze_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
```

```python
sns.countplot(x='gaze', data=gaze_df)
```
